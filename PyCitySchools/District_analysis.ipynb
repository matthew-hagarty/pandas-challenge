## Importing the Data

```python
import pandas as pd
from pathlib import Path

schools_path = Path("Resources/schools_complete.csv")
students_path = Path("Resources/students_complete.csv")

schools_df = pd.read_csv(schools_path)
students_df = pd.read_csv(students_path)

schools_df.head()
```

```python
students_df.head()
```

## District Analysis Code

```python
#number of unique schools
unique_schools = len(schools_df["School ID"].unique())
unique_schools
```

```python
#number of unique students
unique_students = len(students_df["Student ID"].unique())
unique_students
```

```python
#total district budget
district_budget = schools_df["budget"].sum()
district_budget
```

```python
math_avg = students_df["math_score"].mean()
math_avg
```

```python
reading_avg = students_df["reading_score"].mean()
reading_avg
```

```python
math_pass_number = students_df[(students_df["math_score"] >= 70)].count()["Student ID"]
math_pass_pct = math_pass_number / unique_students * 100
math_pass_pct
```

```python
reading_pass_number = students_df[(students_df["reading_score"] >= 70)].count()["Student ID"]
reading_pass_pct = reading_pass_number / unique_students * 100
reading_pass_pct
```

```python
overall_pass_number = students_df[(students_df["math_score"] >= 70) & (students_df["reading_score"] >= 70)].count()["Student ID"]
overall_pass_pct = overall_pass_number / unique_students * 100
overall_pass_pct
```

## District Summary

```python
district_summary_dict = {
    "Total Schools":[unique_schools],
    "Total Students":[unique_students],
    "Total Budget":[district_budget],
    "Average Math Score":[math_avg],
    "Average Reading Score":[reading_avg],
    "% Passing Math":[math_pass_pct],
    "% Passing Reading":[reading_pass_pct],
    "% Overall Passing":[overall_pass_pct]
}

district_summary = pd.DataFrame(district_summary_dict)

# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.2f}".format)

district_summary
```

## Individual School Summaries

```python
schools = schools_df["school_name"]
schools
```

```python
school_type = schools_df["type"]
school_type
```

```python
school_size = schools_df["size"]
school_size
```

```python
school_budget = schools_df["budget"]
school_budget
```

```python
budget_per_student = school_budget / school_size
budget_per_student
```

```python
schools_groupby = students_df.groupby(students_df["school_name"])
school_reading_mean  = schools_groupby["reading_score"].mean()
school_reading_mean
```

```python
school_math_mean  = schools_groupby["math_score"].mean()
school_math_mean
```

```python
all_schools_pass_reading = students_df[students_df["reading_score"] >= 70]
per_school_pass_reading = all_schools_pass_reading.groupby(["school_name"]).size()
per_school_pass_reading
```

```python
all_schools_pass_math = students_df[students_df["math_score"] >= 70]
per_school_pass_math = all_schools_pass_math.groupby(["school_name"]).size()
per_school_pass_math
```

```python
all_schools_pass = students_df[(students_df["reading_score"] >= 70) & (students_df["math_score"] >= 70)]
per_school_pass = all_schools_pass.groupby(["school_name"]).size()
per_school_pass
```

## Individual School Summaries

```python
schools_expanded_df = schools_df.set_index('school_name').drop(columns = ["School ID"])
schools_expanded_df["budget_per_student"] = schools_expanded_df["budget"] / schools_expanded_df['size']
schools_expanded_df["reading_mean"] = school_reading_mean
schools_expanded_df['math_mean'] = school_math_mean
schools_expanded_df['% Passing Reading'] = per_school_pass_reading / schools_expanded_df['size'] * 100
schools_expanded_df['% Passing Math'] = per_school_pass_math / schools_expanded_df['size'] * 100
schools_expanded_df['% Passing Overall'] = per_school_pass / schools_expanded_df['size'] * 100
schools_expanded_df
```

## Top Performing Schools

```python
top_schools = schools_expanded_df.sort_values(["% Passing Overall"], ascending = False).head()
top_schools
```

## Worst Performing Schools

```python
bottom_schools = schools_expanded_df.sort_values(["% Passing Overall"], ascending = True).head()
bottom_schools
```

## By Grade Per School (Math)

```python
ninth_grade = students_df[students_df["grade"] == '9th']
tenth_grade = students_df[students_df["grade"] == '10th']
eleventh_grade = students_df[students_df["grade"] == '11th']
twelfth_grade = students_df[students_df["grade"] == '12th']

ninth_grade_math = ninth_grade.groupby(['school_name'])['math_score'].mean()
tenth_grade_math  = tenth_grade.groupby(['school_name'])['math_score'].mean()
eleventh_grade_math  = eleventh_grade.groupby(['school_name'])['math_score'].mean()
twelfth_grade_math  = twelfth_grade.groupby(['school_name'])['math_score'].mean()

math_scores_per_grade_dict = {
    "9th" : ninth_grade_math,
    "10th" : tenth_grade_math,
    "11th" : eleventh_grade_math,
    "12th" : twelfth_grade_math
}

math_scores_by_grade = pd.DataFrame(math_scores_per_grade_dict)
math_scores_by_grade.index.name = None
math_scores_by_grade
```

## By Grade Per School (Reading)

```python
ninth_grade_reading = ninth_grade.groupby(['school_name'])['reading_score'].mean()
tenth_grade_reading  = tenth_grade.groupby(['school_name'])['reading_score'].mean()
eleventh_grade_reading  = eleventh_grade.groupby(['school_name'])['reading_score'].mean()
twelfth_grade_reading  = twelfth_grade.groupby(['school_name'])['reading_score'].mean()

reading_scores_per_grade_dict = {
    "9th" : ninth_grade_reading,
    "10th" : tenth_grade_reading,
    "11th" : eleventh_grade_reading,
    "12th" : twelfth_grade_reading
}

reading_scores_by_grade = pd.DataFrame(reading_scores_per_grade_dict)
reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

## Scores by School Spending

```python
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spending_df = schools_expanded_df.copy()

school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df['budget_per_student'], spending_bins, labels = spending_labels)
school_spending_df
```

```python
spending_math_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["math_mean"].mean()
spending_reading_scores = school_spending_df.groupby(["Spending Ranges (Per Student)"])["reading_mean"].mean()
spending_passing_math = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Spending Ranges (Per Student)"])["% Passing Overall"].mean()

spending_summary_dict = {
    "Mean Math Score" : spending_math_scores,
    "Mean Reading Score" : spending_reading_scores,
    "% Passing Math" : spending_passing_math,
    "% Passing Reading" : spending_passing_reading,
    "% Passing Overall" : overall_passing_spending
}

spending_summary_df = pd.DataFrame(spending_summary_dict)
spending_summary_df
```

## Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size_df = schools_expanded_df.copy()

school_size_df["School Size (S/M/L)"] = pd.cut(school_size_df["size"], size_bins, labels = size_labels)
school_size_df
```

```python
size_math_scores = school_size_df.groupby(["School Size (S/M/L)"])["math_mean"].mean()
size_reading_scores = school_size_df.groupby(["School Size (S/M/L)"])["reading_mean"].mean()
size_passing_math = school_size_df.groupby(["School Size (S/M/L)"])["% Passing Math"].mean()
size_passing_reading = school_size_df.groupby(["School Size (S/M/L)"])["% Passing Reading"].mean()
size_passing_overall = school_size_df.groupby(["School Size (S/M/L)"])["% Passing Overall"].mean()
```

```python
size_summary_dict = {
    "Average Math Score": size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Passing Overall": size_passing_overall
}

size_summary_df = pd.DataFrame(size_summary_dict)
size_summary_df
```
